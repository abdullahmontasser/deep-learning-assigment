```python
import numpy as np
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)
```

```python
inputs = [1, 2, 3]

h_prev = 0
c_prev = 0

W_f, U_f, b_f = 0.5, 0.1, 0.1
W_i, U_i, b_i = 0.6, 0.1, 0.1
W_c, U_c, b_c = 0.7, 0.1, 0.1
W_o, U_o, b_o = 0.8, 0.1, 0.1

W_y = 1.0
b_y = 0.0
```

```python
print(f"{'Time Step':<10} | {'Input':<5} | {'Forget (f)':<10} | {'Input (i)':<10} | {'Candidate':<12} | {'Cell State':<12} | {'Output (o)':<10} | {'Hidden (h)':<10}")
print("-" * 105)

for t, x_t in enumerate(inputs):
    f_t = sigmoid(W_f * x_t + U_f * h_prev + b_f)
    i_t = sigmoid(W_i * x_t + U_i * h_prev + b_i)
    c_tilde = tanh(W_c * x_t + U_c * h_prev + b_c)
    c_t = f_t * c_prev + i_t * c_tilde
    o_t = sigmoid(W_o * x_t + U_o * h_prev + b_o)
    h_t = o_t * tanh(c_t)

    print(f"{t+1:<10} | {x_t:<5} | {f_t:<10.4f} | {i_t:<10.4f} | {c_tilde:<12.4f} | {c_t:<12.4f} | {o_t:<10.4f} | {h_t:<10.4f}")

    h_prev = h_t
    c_prev = c_t
```

```python
y_hat = W_y * h_prev + b_y

print("-" * 105)
print(f"Final Hidden State (h_3): {h_prev:.4f}")
print(f"Final Prediction (y_hat): {y_hat:.4f}")
```
